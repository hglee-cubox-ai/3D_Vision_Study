```python
import cv2
import numpy as np
import IPython

def imshow(img):
    _,ret = cv2.imencode('.jpg', img) 
    i = IPython.display.Image(data=ret)
    IPython.display.display(i)
```

6ê°•. Direct Linear Transform

```python
img_src = cv2.imread('./samples/book.jpg')
imshow(img_src)
```

```python
pts_src = np.array([
    [283, 229], # left up
    [479, 342], # right up
    [283, 603], # right down
    [61, 428]   # left down
])

pts_dst = np.array([
    [0, 0],
    [200, 0],
    [200, 400],
    [0, 400],
])
dst_width = 200
dst_height = 400

h, status = cv2.findHomography(pts_src, pts_dst)
```

```python
im_dst = cv2.warpPerspective(img_src, h, (dst_width, dst_height))

imshow(im_dst)
```
